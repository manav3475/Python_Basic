### Python Variables, Naming & Dynamic Typing

In Python you don't declare a type before a variable — you just give it a name and assign a value, and Python figures out the rest. Today I'll cover how assignment works, the snake_case naming convention Python developers follow, the reserved keywords you're not allowed to use as names, and the big one — dynamic typing, where a variable can hold a number now and a string a second later. Let's go!

#### 1. Assignment with `=`

The `=` sign assigns the value on the right to the name on the left. No type keyword needed.

```python
name = "Manav"
age = 25
height = 5.9

print(name, age, height)
```

#### Multiple assignment

Python lets you assign several variables in one line — either the same value to many names, or unpack several values at once.

```python
# Same value to several names
x = y = z = 0
print(x, y, z)

# Unpack several values at once
a, b, c = 1, 2, 3
print(a, b, c)
```

This unpacking is also the clean way to swap two variables — no temporary variable needed.

```python
p, q = 10, 20
p, q = q, p        # swap in one line
print("p =", p, "| q =", q)
```

#### 2. The snake_case convention

Python's style guide (PEP 8) says variable and function names should be lowercase with words separated by underscores. That's snake_case. It's not enforced by the language, but every Python codebase follows it.

```python
# Good - snake_case
first_name = "Manav"
total_price = 199.99
is_active = True

# Works, but not Pythonic style
firstName = "Manav"   # camelCase - avoid in Python
TotalPrice = 199.99   # PascalCase - reserved for class names

print(first_name, total_price, is_active)
```

#### The naming rules Python *does* enforce

Style is optional, but these rules are hard requirements — break them and the code won't run:

- A name must start with a letter or underscore, never a digit.
- It can only contain letters, digits, and underscores.
- No spaces, no hyphens, no other symbols.

```python
# These are all valid names
_private = 1
user2 = 2
my_var_3 = 3
print(_private, user2, my_var_3)
```

Now watch an invalid name fail — a variable can't start with a number.

```python
2fast = "nope"   # starts with a digit -> SyntaxError
```

#### 3. Reserved keywords

Python has a set of words that are part of the language itself — like `if`, `for`, `class`, `return`. You can't use them as variable names. Here's the full list, pulled from Python directly.

```python
import keyword

print("Total keywords:", len(keyword.kwlist))
print(keyword.kwlist)
```

Try to use one as a name and Python rejects it immediately.

```python
class = "my class"   # 'class' is reserved -> SyntaxError
```

Tip: if you ever want a name that clashes with a keyword, the convention is to add a trailing underscore — `class_` instead of `class`.

```python
class_ = "this works fine"
print(class_)
```

#### 4. Dynamic typing

This is what sets Python apart from Java or C++. You never declare a type. The same variable can hold different types over its life, and Python tracks the current type for you. The built-in `type()` shows what a variable holds right now.

```python
data = 42
print(data, "->", type(data))

data = "now I'm a string"
print(data, "->", type(data))

data = [1, 2, 3]
print(data, "->", type(data))

data = 3.14
print(data, "->", type(data))
```

The name `data` never changed, but its type went from int to str to list to float. Python re-points the name to whatever new value you assign. That flexibility is powerful — but it also means a variable's type can surprise you if you're not careful, which is why `type()` is handy while debugging.

#### Quick recap

- Assign with `=`; no type declaration needed.
- Assign many at once: `x = y = 0` or `a, b = 1, 2`, and swap with `a, b = b, a`.
- Use snake_case for variables and functions; names must start with a letter or underscore and contain only letters, digits, and underscores.
- Reserved keywords can't be used as names — check them with `keyword.kwlist`.
- Python is dynamically typed: a variable can hold any type, and `type()` tells you the current one.

That's the foundation of every Python program. Next time you name a variable, you'll know exactly what Python allows.
