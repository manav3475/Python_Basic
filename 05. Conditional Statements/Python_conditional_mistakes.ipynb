### Common Conditional Mistakes in Python

Conditionals look simple, but they're where beginners lose the most hours to silent bugs — code that runs without error but does the wrong thing. Today I'll walk through the most common mistakes with `if` statements, show you exactly what goes wrong, and give you the clean fix for each. Learn these once and you'll debug your own code far faster. Let's go!

#### Mistake 1: using `=` instead of `==`

A single `=` assigns; a double `==` compares. Mixing them up is the classic. The good news: Python catches this one with a SyntaxError, so it can't silently pass. Let's see it.

```python
age = 18

# WRONG: single = inside an if
if age = 18:
    print("adult")
```

The fix is simply `==` for the comparison.

```python
age = 18

if age == 18:      # correct: compare with ==
    print("adult")
```

#### Mistake 2: wrong comparison for a range

Beginners often write conditions that can never both be true, or forget Python supports chaining. You can't check a range with `if 0 < x` on its own and hope it means between values.

```python
x = 5

# clumsy and easy to get wrong
if x > 0 and x < 10:
    print("in range (works, but verbose)")

# clean, Pythonic chaining
if 0 < x < 10:
    print("in range (clean)")
```

#### Mistake 3: misunderstanding truthy / falsy

`if my_list:` checks whether the list is *non-empty*, not whether it exists. Values like `0`, `""`, `[]`, `{}`, and `None` are all falsy. Forgetting this leads to surprising skips.

```python
items = []

if items:
    print("has items")
else:
    print("empty list is FALSY -> we land here")

# also falsy:
for v in [0, "", None, [], {}]:
    print(repr(v), "->", bool(v))
```

#### Mistake 4: comparing to True / False explicitly

Writing `if is_active == True:` is redundant and can even bite you. Just use the value directly.

```python
is_active = True

# redundant
if is_active == True:
    print("verbose")

# clean and correct
if is_active:
    print("Pythonic")
```

#### Mistake 5: checking None with == instead of is

`None` is a unique object; the correct, readable test is `is None`, not `== None`.

```python
value = None

if value == None:      # works, but not recommended
    print("using == (avoid)")

if value is None:      # correct and Pythonic
    print("using is (correct)")
```

#### Mistake 6: forgetting elif — stacking separate ifs

Using several independent `if`s when you mean *one* decision can let multiple blocks run. Use `elif` so exactly one branch fires.

```python
score = 95

# BUG: separate ifs -> two lines print!
print("--- separate ifs ---")
if score >= 90:
    print("Grade A")
if score >= 80:
    print("Grade B")   # this also runs!

# FIX: elif -> only the first true branch runs
print("--- elif chain ---")
if score >= 90:
    print("Grade A")
elif score >= 80:
    print("Grade B")
```

#### Mistake 7: indentation errors in blocks

Python decides what's inside the `if` by indentation. A mis-indented line either errors or silently runs when it shouldn't. Here the print always runs because it's outside the if.

```python
logged_in = False

# the print is NOT indented under the if, so it always runs
if logged_in:
    pass
print("Welcome!")   # oops - shows even when logged_in is False
```

The fix is to indent the line so it belongs to the block.

```python
logged_in = False

if logged_in:
    print("Welcome!")   # now only runs when logged_in is True
else:
    print("Please log in")
```

#### Quick recap

- Use `==` to compare, `=` only to assign.
- Chain ranges the Pythonic way: `0 < x < 10`.
- Empty values (`0`, `""`, `[]`, `{}`, `None`) are falsy — `if my_list:` means "non-empty".
- Don't write `== True`; just use the value. Test None with `is None`.
- Use `elif` for one decision so only one branch runs.
- Watch your indentation — it decides what's inside the block.
