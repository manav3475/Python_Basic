### Python match-case Statement (Structural Pattern Matching)

If you've come from another language you might miss the classic `switch` statement — and for a long time Python didn't have one. Since Python 3.10 we have something better: `match` / `case`, also called structural pattern matching. It's a clean way to compare one value against many possible patterns without a long `if-elif-elif` ladder. Today we'll build a menu selector, match multiple values, use a catch-all, add conditions, and even unpack data inside a case. Let's go!

#### 1. The basic match-case

You `match` a value, then list `case` patterns. Python runs the first one that matches. The `_` case is the catch-all (like `else`).

```python
command = "start"

match command:
    case "start":
        print("Starting the engine...")
    case "stop":
        print("Stopping.")
    case _:
        print("Unknown command")
```

That `_` at the bottom is the wildcard — it matches anything that didn't match above. Without it, a value with no matching case simply does nothing (no error).

#### 2. Matching several values with one case

Use the pipe `|` to let one case match multiple options — handy for grouping.

```python
day = "Sun"

match day:
    case "Sat" | "Sun":
        print("Weekend!")
    case "Mon" | "Tue" | "Wed" | "Thu" | "Fri":
        print("Weekday.")
    case _:
        print("Not a valid day")
```

#### 3. A practical menu / command dispatcher

This is where match-case shines — turning a user's choice into an action.

```python
def run_menu(choice):
    match choice:
        case "1":
            return "New Game"
        case "2":
            return "Load Game"
        case "3":
            return "Quit"
        case _:
            return "Invalid choice, try again"

print(run_menu("1"))
print(run_menu("3"))
print(run_menu("9"))
```

#### 4. Guards — add a condition to a case

You can attach an `if` to a case. The case only matches when the pattern fits AND the condition is true. This is called a guard.

```python
num = 7

match num:
    case n if n < 0:
        print("Negative")
    case n if n == 0:
        print("Zero")
    case n if n % 2 == 0:
        print("Positive even")
    case _:
        print("Positive odd")
```

#### 5. Unpacking structures (the real power)

match-case can look *inside* data and pull values out. Here we match on the shape of a coordinate tuple and capture the parts we need.

```python
point = (0, 5)

match point:
    case (0, 0):
        print("Origin")
    case (0, y):
        print(f"On the Y-axis at y={y}")
    case (x, 0):
        print(f"On the X-axis at x={x}")
    case (x, y):
        print(f"Somewhere at ({x}, {y})")
```

Notice how `(0, y)` matched our point `(0, 5)` and captured `y = 5` automatically. That pattern matching on structure is something a plain `switch` in other languages can't do.

#### 6. When to use match-case vs if-elif

- Reach for **match-case** when you're checking one value against many distinct patterns, or unpacking structured data.
- Stick with **if-elif** for ranges and unrelated boolean conditions (`if temp > 30 and is_sunny`).
- Remember: match-case needs Python 3.10 or newer.

#### Quick recap

- `match value:` then `case pattern:` runs the first matching case; `case _:` is the catch-all.
- Combine options with `|`: `case "Sat" | "Sun":`.
- Add a guard with `if`: `case n if n % 2 == 0:`.
- Unpack structures directly: `case (0, y):` captures `y`.
- Great for menus and command dispatchers; needs Python 3.10+.
