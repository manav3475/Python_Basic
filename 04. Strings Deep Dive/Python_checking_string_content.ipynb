### Python: Checking String Content

Very often you need to *ask questions* about a string: does this filename end with .txt? did the user actually type a number? is this box empty? Python strings come with a set of built-in check methods that answer with a clean `True` or `False`. Today we'll cover `startswith`, `endswith`, and the `is...` family — `isdigit`, `isalpha`, `isspace`, and a couple of friends. Let's go!

#### 1. startswith() and endswith()

These check whether a string begins or ends with a given piece of text. Great for filenames, URLs, and prefixes.

```python
filename = "report.pdf"

print(filename.endswith(".pdf"))      # True
print(filename.endswith(".txt"))      # False
print(filename.startswith("report"))  # True

url = "https://youtube.com"
print(url.startswith("https"))        # True -> is it secure?
```

Both are case sensitive, and both accept a tuple to check several options at once.

```python
name = "Manav.PNG"

# check for multiple endings in one call
print(name.lower().endswith((".png", ".jpg", ".jpeg")))   # True
```

#### 2. isdigit() — is every character a digit?

This is the classic way to check whether input looks like a whole number *before* you convert it, so you don't crash on bad input.

```python
print("12345".isdigit())    # True
print("12a45".isdigit())    # False -> has a letter
print("".isdigit())         # False -> empty string
print("3.14".isdigit())     # False -> the dot is not a digit
```

A safe pattern: check before you cast.

```python
user_input = "42"

if user_input.isdigit():
    age = int(user_input)
    print("Next year you will be", age + 1)
else:
    print("Please enter a valid number")
```

#### 3. isalpha() — only letters?

True when every character is a letter (no digits, spaces, or symbols). Handy for validating names.

```python
print("Manav".isalpha())     # True
print("Manav123".isalpha())  # False -> has digits
print("Manav Shah".isalpha())# False -> the space counts!
print("hello".isalpha())     # True
```

#### 4. isspace() — is it all whitespace?

True when the string is non-empty and made only of spaces, tabs, or new lines. Useful for catching "blank" input that looks empty but isn't.

```python
print("   ".isspace())    # True  -> just spaces
print("\t\n".isspace())   # True  -> tab + newline
print("a b".isspace())    # False -> has letters
print("".isspace())       # False -> empty string is NOT whitespace
```

#### 5. A few more of the family

The same idea extends to several handy checks.

```python
print("hello".islower())     # True  -> all lowercase
print("HELLO".isupper())     # True  -> all uppercase
print("Hello World".istitle())# True -> Each Word Capitalised
print("user_2".isalnum())    # False -> underscore is not a letter/digit
print("user2".isalnum())     # True  -> letters + digits only
```

#### 6. Putting it together — a tiny validator

These checks shine when you combine them to validate real input.

```python
def check_username(u):
    if u.isspace() or u == "":
        return "Username cannot be blank"
    if not u.isalnum():
        return "Only letters and numbers allowed"
    return "Username looks good!"

print(check_username("manav25"))
print(check_username("manav 25"))
print(check_username("   "))
```

#### Quick recap

- `startswith()` / `endswith()` check a prefix or suffix; both are case sensitive and accept a tuple of options.
- `isdigit()` -> all digits (check before converting to int).
- `isalpha()` -> all letters (a space makes it False).
- `isspace()` -> only whitespace, and empty string is False.
- Friends: `islower`, `isupper`, `istitle`, `isalnum`.
- Combine them to validate usernames, filenames, and user input.
