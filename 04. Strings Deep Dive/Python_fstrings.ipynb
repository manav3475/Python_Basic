### Python String Formatting with f-strings

Sooner or later every program needs to build text out of variables — a greeting with someone's name, a price with two decimals, a neat table of results. The cleanest way to do this in modern Python is the **f-string**. You put an `f` before the quotes, then drop variables straight into the text inside `{ }`. Today we'll cover the basics, running expressions inside the braces, formatting numbers, and lining things up in columns. Let's go!

#### 1. The basics — drop a variable into text

Put `f` right before the opening quote, then write `{variable}` anywhere inside. Python swaps in the value for you.

```python
name = "Manav"
age = 25

print(f"Hi, I am {name} and I am {age} years old.")
```

Before f-strings, people glued text together with `+` and had to convert numbers by hand. f-strings are shorter and far easier to read.

```python
# the old, clumsy way
print("Hi, I am " + name + " and I am " + str(age) + " years old.")

# the f-string way - same result, much cleaner
print(f"Hi, I am {name} and I am {age} years old.")
```

#### 2. Run expressions inside the braces

The braces aren't limited to plain variables — you can put any Python expression in them, and it runs right there.

```python
a = 8
b = 5

print(f"{a} + {b} = {a + b}")
print(f"{a} is even? {a % 2 == 0}")
print(f"Uppercase name: {name.upper()}")
```

#### 3. Formatting numbers

After the value, add a colon `:` and a format spec. The most common one is rounding a float to a fixed number of decimals with `.2f` (2 digits after the point).

```python
price = 199.5
pi = 3.14159265

print(f"Price: {price:.2f}")      # 199.50  -> 2 decimals
print(f"Pi rounded: {pi:.3f}")    # 3.142   -> 3 decimals

# thousands separator with a comma
big = 1000000
print(f"Total: {big:,}")          # 1,000,000
```

You can also show percentages and pad numbers with leading zeros.

```python
ratio = 0.8734
print(f"Score: {ratio:.1%}")      # 87.3%  -> percent format

num = 7
print(f"ID: {num:03d}")           # 007    -> pad to 3 digits with zeros
```

#### 4. Alignment — building neat columns

Inside the format spec, `<` left-aligns, `>` right-aligns, and `^` centers, followed by a width. This is how you line text up into tidy columns.

```python
items = [("Apple", 30), ("Banana", 5), ("Watermelon", 45)]

for fruit, qty in items:
    # fruit left-aligned in 12 spaces, qty right-aligned in 4
    print(f"{fruit:<12}{qty:>4}")
```

#### 5. Handy trick: the `=` for quick debugging

Putting `=` right after a variable inside the braces prints both its name and value — perfect for quick debugging (Python 3.8+).

```python
x = 42
y = 10
print(f"{x=}")        # x=42
print(f"{x + y = }")  # x + y = 52
```

#### Quick recap

- Start with `f"..."` and put variables inside `{ }`.
- Any expression works inside the braces: `{a + b}`, `{name.upper()}`.
- Format numbers after a colon: `{price:.2f}`, `{big:,}`, `{ratio:.1%}`, `{num:03d}`.
- Align with `<` `>` `^` and a width to build columns: `{fruit:<12}`.
- `{x=}` prints the name and value together — great for debugging.
