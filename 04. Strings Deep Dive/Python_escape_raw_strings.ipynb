### Python Escape Characters & Raw Strings

What if you want a new line inside a string, or a tab, or a quote mark that's the same as the quotes wrapping your string? You can't just type them — you need **escape characters**, little backslash codes that stand for special characters. And when backslashes get in your way (think Windows file paths), **raw strings** switch the whole system off. Let's go!

#### 1. New line `\n` and tab `\t`

A backslash `\` followed by a letter is an escape sequence — it doesn't print literally, it does something. `\n` starts a new line; `\t` inserts a tab.

```python
print("Line one\nLine two")
print("Name:\tManav")
print("A\tB\tC")
```

#### 2. Quotes inside quotes

If your string is wrapped in double quotes and you want a double quote inside, escape it with `\"`. Same idea for single quotes inside single quotes.

```python
print("She said \"Hello\" to me.")
print('It\'s a sunny day.')

# easier alternative: use the OTHER quote type on the outside
print('She said "Hello" to me.')
print("It's a sunny day.")
```

#### 3. A literal backslash `\\`

Since `\` starts an escape, to print an actual backslash you need two of them: `\\`.

```python
print("Back\\slash")        # Back\slash
print("C:\\Users\\Manav")   # C:\Users\Manav
```

#### 4. The backslash headache — file paths

Look what goes wrong with a Windows path. The `\n` in `\name` and `\t` in `\test` secretly become a newline and a tab!

```python
path = "C:\name\test.txt"
print(path)     # notice \n became a new line and \t became a tab!
```

#### 5. Raw strings `r"..."` — turn escapes OFF

Put an `r` before the quotes and Python treats every backslash literally — no escape processing at all. This is the clean fix for file paths and regular-expression patterns.

```python
path = r"C:\name\test.txt"
print(path)     # C:\name\test.txt  -> exactly as written

print(r"New\nline stays literal")   # the \n is NOT a new line here
```

One caveat: a raw string can't *end* with a single backslash, because the backslash would escape the closing quote. That's a rare edge case, but good to know.

#### 6. Bonus: multi-line strings with triple quotes

When you want several lines without sprinkling `\n` everywhere, wrap the text in triple quotes and just press Enter.

```python
text = """Dear Manav,
Welcome to the course.
Happy coding!"""

print(text)
```

#### Quick recap

- Escape sequences start with `\`: `\n` new line, `\t` tab.
- Escape a quote that matches your wrapper with `\"` or `\'` — or just use the other quote type.
- A literal backslash is `\\` (two of them).
- Raw strings `r"..."` turn escapes OFF — perfect for Windows paths and regex.
- Triple quotes `"""..."""` make clean multi-line strings.
