### Python `print()` — Beyond the Basics

Everyone's first Python line is `print("Hello, World!")` — but `print()` can do a lot more than dump text on a line. It takes multiple arguments, lets you control the separator between them with `sep`, change what goes at the end with `end`, and even force output to appear instantly with `flush`. Today I'll show all four, then build a clean formatted line using them. Let's go!

#### 1. Multiple arguments

`print()` accepts as many values as you want, separated by commas. It prints them all on one line — and notice you can mix types freely, no conversion needed.

```python
name = "Manav"
age = 25

print("Name:", name, "Age:", age)
print("Mixing types:", 1, 2.5, True, "text")
```

By default Python puts a single space between each argument. That space is actually controlled by a parameter — which is our next stop.

#### 2. The `sep` parameter

`sep` sets what goes *between* the arguments. The default is a single space, but you can make it anything — a comma, a dash, a newline, even nothing.

```python
# Default separator is a space
print("a", "b", "c")

# Custom separators
print("a", "b", "c", sep=", ")
print("2024", "01", "15", sep="-")
print("no", "gaps", sep="")
```

A neat trick — use `sep="\n"` to put each argument on its own line with a single print call.

```python
print("Line one", "Line two", "Line three", sep="\n")
```

#### 3. The `end` parameter

Every `print()` ends with a newline by default — that's why each print starts on a fresh line. `end` changes that. Set it to a space and the next print continues on the same line.

```python
# Default: each print goes on its own line
print("First")
print("Second")
```

```python
# end=' ' keeps them on the same line
print("First", end=" ")
print("Second", end=" ")
print("Third")
```

This is perfect for building a line piece by piece — like printing numbers in a row instead of a column.

```python
for i in range(1, 6):
    print(i, end=" ")
print()   # a bare print() just prints the newline to finish the line
```

#### 4. The `flush` parameter

Python often *buffers* output — it collects text and writes it in chunks instead of instantly. Usually you don't notice. But for live output like a progress indicator or a countdown, buffering can make text appear late. `flush=True` forces it out immediately.

```python
import time

# Without flush, on some setups this could appear all at once at the end.
# flush=True guarantees each dot shows the moment it's printed.
print("Loading", end="")
for _ in range(5):
    print(".", end="", flush=True)
    time.sleep(0.3)
print(" done!")
```

In a notebook you may not see the delay clearly, but in a real terminal script the difference is obvious — with `flush=True` the dots appear one by one; without it they can all pop up together at the end. Rule of thumb: use `flush=True` whenever you're showing progress in real time.

#### 5. Putting it all together — a formatted line

Now let's combine the parameters to print a clean, formatted output line without any string concatenation.

```python
product = "Toothpaste"
brand = "Colgate"
size = "100ml"
price = 85

# sep joins the fields with ' | ', end adds a custom closing
print(brand, product, size, f"Rs.{price}", sep=" | ", end="  <-- in stock\n")
```

One call, four fields, a clean separator, and a custom ending — no `+` signs and no manual spacing. For comparison, here's the same output built with an f-string, which is the other common approach.

```python
print(f"{brand} | {product} | {size} | Rs.{price}  <-- in stock")
```

Both give the same result. Use `sep`/`end` when you have a list of values to join; reach for an f-string when you want fine control over the exact layout.

#### Quick recap

- `print()` takes multiple comma-separated arguments and prints them on one line.
- `sep` controls what goes between arguments (default is a space).
- `end` controls what goes at the end (default is a newline `\n`).
- `flush=True` forces output to appear immediately — use it for live progress output.
- Combine them to format a line in a single call, or use an f-string for precise layouts.

The humble `print()` is more capable than it looks — now you know every lever it gives you.
