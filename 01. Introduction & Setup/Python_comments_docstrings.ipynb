### Python Comments & Docstrings

Code tells the computer what to do — but comments tell *humans* what's going on. Python gives you three ways to document your code: the single-line `#` comment, block comments, and triple-quoted docstrings. Today I'll show you when to use each one, how docstrings are different from regular comments, and how Python can read your docstrings back to you at runtime. Let's go!

#### 1. The single-line comment `#`

Anything after a `#` on a line is ignored by Python. It's purely for the reader.

```python
# This whole line is a comment - Python skips it
print("Hello, World!")   # this part after the # is a comment too
```

The first line never runs. On the second line, Python runs the `print` and ignores everything from the `#` onward. That second style is called an *inline comment*.

#### 2. Block comments

Python has no dedicated multi-line comment symbol. A block comment is just several `#` lines stacked together. Use it to explain the section of code that follows.

```python
# Calculate the average of three test scores.
# We add them up first, then divide by 3.
# The result is stored in `average`.
s1, s2, s3 = 80, 90, 70
average = (s1 + s2 + s3) / 3
print("Average:", average)
```

#### The 'comment out' trick

Comments aren't just for notes. You can put a `#` in front of a line to temporarily disable it without deleting it — handy while debugging.

```python
print("This runs")
# print("This is commented out, so it never runs")
print("This runs too")
```

#### 3. Triple-quoted strings

You'll often see triple quotes `""" """` used to span multiple lines. Strictly speaking these are *strings*, not comments — but when a string isn't assigned to anything, Python just creates it and throws it away, so people sometimes use them as multi-line notes.

```python
"""
This is a triple-quoted string sitting on its own.
It spans multiple lines. Python evaluates it but since
we don't store it, nothing happens - it acts like a comment.
"""
print("Code after the triple-quoted block still runs")
```

But there's one special place where a triple-quoted string is *not* thrown away — and that's where docstrings come in.

#### 4. Docstrings — documentation Python remembers

A docstring is a triple-quoted string placed as the **first line inside** a function, class, or module. Python attaches it to that object so it can be read later. This is the real difference: a `#` comment is invisible at runtime, but a docstring is stored and retrievable.

```python
def add(a, b):
    """Return the sum of a and b."""
    return a + b

print(add(2, 3))
```

#### Reading a docstring back

Because Python stores the docstring, you can access it two ways — through the `__doc__` attribute, or with the built-in `help()` function.

```python
print(add.__doc__)
```

```python
help(add)
```

#### A fuller docstring

For anything non-trivial, a multi-line docstring describes what the function does, its arguments, and what it returns. This is the standard format most teams follow.

```python
def divide(a, b):
    """
    Divide a by b and return the result.

    Args:
        a (float): the numerator
        b (float): the denominator (must not be zero)

    Returns:
        float: the result of a / b
    """
    return a / b

help(divide)
```

#### Docstrings work on classes too

The same rule applies — first line inside the class or method.

```python
class Calculator:
    """A tiny calculator class for demonstration."""

    def square(self, n):
        """Return n multiplied by itself."""
        return n * n

help(Calculator)
```

#### Comment on the *why*, not the *what*

A common beginner habit is commenting things the code already says. Good comments explain reasoning the code can't show on its own.

```python
x = 0   # start on the first page

# Bad - just repeats the code
x = x + 1   # add 1 to x

# Better - explains the reason
x = x + 1   # move to the next page; pages are 1-indexed in the API

print("x is now", x)
```

#### Quick recap

- `#` — single-line and inline comments, ignored by Python at runtime.
- Stacked `#` lines — block comments for a section of code.
- Triple quotes — multi-line strings; act like comments when unassigned.
- Docstrings — triple-quoted strings as the first line of a function, class, or module; Python stores them and you read them with `help()` or `__doc__`.
- Comment the *why*, not the *what*.

Get in the habit early and your future self — and your teammates — will thank you.
