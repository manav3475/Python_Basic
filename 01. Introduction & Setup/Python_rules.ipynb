### Python Syntax & Indentation Rules

In most programming languages like Java or C++, you use curly braces to define blocks of code. But Python? Python throws all that away and uses something surprisingly simple — whitespace. And if you get it wrong, your code just won't run. Today I'll show you exactly how Python's syntax and indentation works, and we'll even break the code on purpose to see the famous IndentationError. Let's go!

#### Java

```java
public class HelloWorld {

    public static void main(String[] args) {
        System.out.println("Hello, World!");
    }
}
```

Notice the curly braces `{ }` — that's how Java knows where a block starts and ends. The indentation is just for humans; Java ignores it.

### Python

Same program in Python. No class, no braces, no semicolons — just one line.

````python
print("Hello, World!")
````

#### Why indentation matters

In Python, indentation is not decoration — it *is* the syntax. The spaces in front of a line tell Python which lines belong together as a block. Where Java uses `{ }`, Python uses the indent level.

#### The two rules to remember

1. A line that opens a block ends with a colon `:` (like `if`, `for`, `while`, `def`).
2. Every line inside that block must be indented by the same amount — the standard is 4 spaces.

````python
age = 20

if age >= 18:
    print("You are an adult")   # indented -> belongs to the if
    print("You can vote")       # same indent -> same block

print("This line runs no matter what")   # not indented -> outside the if
````

The two indented lines run only when the condition is true. The last line has no indent, so it's outside the `if` and always runs.

#### Same idea inside a function

A `def` opens a block too. Everything that belongs to the function is indented underneath it.

````python
def greet(name):
    message = "Hello, " + name   # inside the function
    return message               # still inside the function

print(greet("Manav"))
````

#### Now let's break it on purpose

This is the moment from the intro. The code below looks fine at a glance, but the second print has an extra space in front of it that doesn't line up with anything. Run it and watch Python complain.

````python
if True:
    print("first line")
     print("second line has one extra space")
````

There it is — the famous **IndentationError: unexpected indent**. Python expected the two lines in the block to line up, and they didn't. In a brace language this would have been fine; in Python it stops everything.

#### A second way to trigger it

The opposite mistake — forgetting to indent after a colon. Python opens a block, expects an indented line next, and finds none.

````python
if True:
print("I forgot to indent this line")
````

This one raises **IndentationError: expected an indented block**. The colon promised a block, but nothing was indented underneath it.

#### The fix

Just make the indentation consistent — 4 spaces for everything inside the block. Here's the corrected version.

````python
if True:
    print("first line")
    print("second line lines up now")
````

#### Spaces vs tabs — pick one

Python lets you use spaces or tabs, but you can't mix them in the same block. The community standard (PEP 8) is 4 spaces. Set your editor to insert spaces when you press Tab and you'll never think about this again.

#### Nesting — blocks inside blocks

Each deeper level adds another 4 spaces. Here a `for` loop contains an `if`, and the `if` contains a `print`.

````python
for i in range(5):
    if i % 2 == 0:
        print(i, "is even")
    else:
        print(i, "is odd")
````

#### Quick recap

- Blocks open with a colon `:` and are defined by indentation, not braces.
- Use 4 spaces per level, and keep it consistent.
- Too much indent → `unexpected indent`. Missing indent after a colon → `expected an indented block`.
- Don't mix tabs and spaces.

That's the whole rule set. Get the indentation right and Python reads almost like plain English.
