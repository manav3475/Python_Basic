### Python `input()` & Type Casting

So far our programs have only talked *to* the user. Now let's let the user talk back. The `input()` function reads what someone types — but there's a catch that trips up almost every beginner: `input()` *always* gives you a string, even when the person types a number. Today I'll show how to read a name and an age, why the age needs converting, and how to cast it with `int()`. Let's go!

#### 1. Reading input

`input()` shows a prompt, waits for the user to type something and press Enter, then hands you back what they typed. Here we read a name.

> Note: in a notebook, running this cell opens a text box for you to type into. The examples below use fixed values so the notebook runs top-to-bottom without stopping — but the commented `input()` line is what you'd use for real.

```python
# Real version (uncomment to try interactively):
# name = input("Enter your name: ")

# For a notebook that runs straight through, we'll simulate the typed value:
name = "Manav"
print("Hello,", name)
```

#### 2. The gotcha — input() always returns a string

Here's the trap. Let's say the user types their age. Even though `25` looks like a number, `input()` returns it as the *string* `"25"`. Watch what `type()` reports.

```python
# Simulating: age_text = input("Enter your age: ")
age_text = "25"          # this is what input() would return - a string

print("Value:", age_text)
print("Type: ", type(age_text))
```

It's a `str`, not an `int`. And that causes real problems the moment you try to do math with it.

```python
age_text = "25"

# Trying to add a number to a string fails
print(age_text + 5)
```

That's a **TypeError** — Python won't add a number to a string. Even the `+` that *does* work between two strings won't help you, because `"25" + "5"` gives `"255"`, not `30`. We need a real integer.

```python
# Two strings 'add' by joining, not by math
print("25" + "5")
```

#### 3. The fix — cast with int()

`int()` converts a string of digits into an actual integer. Once converted, math works as expected.

```python
age_text = "25"           # string from input()
age = int(age_text)       # cast to integer

print("Value:", age)
print("Type: ", type(age))
print("Next year you'll be", age + 1)
```

#### Casting directly on input()

You'll usually see the conversion wrapped straight around `input()` in one line, so the value is an integer from the start.

```python
# Real one-liner version:
# age = int(input("Enter your age: "))

# Simulated:
age = int("30")
print("In 5 years you'll be", age + 5)
```

#### Other casts — float(), str()

`int()` isn't the only one. Use `float()` for decimals, and `str()` to go the other way and turn a number back into text.

```python
height = float("5.9")     # string -> float
print(height, type(height))

count = 42
count_text = str(count)   # int -> string
print(count_text, type(count_text))
```

#### One caution — casting can fail

`int()` only works if the string actually looks like a whole number. Hand it something that isn't, and it raises a ValueError.

```python
print(int("hello"))   # not a number -> ValueError
```

In real programs you'd guard against this with a `try / except` block, so a bad entry doesn't crash the whole thing.

```python
user_value = "hello"

try:
    age = int(user_value)
    print("Got age:", age)
except ValueError:
    print("That wasn't a valid number - please enter digits only.")
```

#### 4. Full example — read name and age

Putting it together the way you'd actually write it. The `input()` lines are shown commented, with simulated values right below so this notebook runs end to end.

```python
# --- Real interactive version ---
# name = input("Enter your name: ")
# age  = int(input("Enter your age: "))

# --- Simulated for the notebook ---
name = "Manav"
age  = int("25")

print(f"Hi {name}, you are {age} years old.")
print(f"Ten years from now, you'll be {age + 10}.")
```

#### Quick recap

- `input(prompt)` reads a line the user types and returns it.
- It **always** returns a string — even `"25"` is text, not a number.
- Doing math on that string raises a `TypeError`; cast it first.
- `int()` converts to a whole number, `float()` to a decimal, `str()` back to text.
- The common pattern is `int(input(...))`, and wrap it in `try / except ValueError` to handle bad input.

Remember the golden rule — *input is always a string until you convert it.* That one line saves beginners hours of confusion.
