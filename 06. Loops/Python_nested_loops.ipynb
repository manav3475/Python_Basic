### Nested Loops in Python

A nested loop is simply a loop inside another loop. For every single turn of the outer loop, the inner loop runs all the way through. This is how you build grids, tables, and patterns. Today we'll build a multiplication table and a star pattern, and understand exactly how the two loops work together. Let's go!

#### 1. The core idea

The inner loop completes fully for each step of the outer loop. Watch the order of the output.

```python
for i in range(1, 4):          # outer
    for j in range(1, 4):      # inner
        print(f"i={i}, j={j}")
    print("---")               # runs once per outer step
```

See how `j` runs 1,2,3 completely for each value of `i`? That's the key: inner finishes before the outer moves on.

#### 2. A multiplication table

The classic use. The outer loop picks the row, the inner loop fills the columns.

```python
for i in range(1, 6):
    for j in range(1, 6):
        print(f"{i*j:3}", end="")   # width 3 keeps columns aligned
    print()                          # newline after each row
```

#### 3. A star pattern (right triangle)

Patterns are where nested loops really click. The row number decides how many stars to print.

```python
for row in range(1, 6):
    for star in range(row):
        print("*", end="")
    print()
```

#### 4. A number triangle

Same structure, printing the column number instead of a star.

```python
for row in range(1, 6):
    for num in range(1, row + 1):
        print(num, end="")
    print()
```

#### 5. A quick word on performance

Nested loops multiply the work: an outer of 100 and an inner of 100 means 10,000 iterations. That's fine for small ranges, but keep it in mind before nesting big loops.

#### Quick recap

- A nested loop is a loop inside a loop.
- The inner loop runs fully for every single step of the outer loop.
- Outer loop = rows, inner loop = columns is the mental model for tables and patterns.
- Use `end=""` to stay on the same line and `print()` to move to the next row.
- Nested loops multiply iterations — watch the size.
