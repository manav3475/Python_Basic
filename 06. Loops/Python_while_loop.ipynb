### Python while Loop Basics

A `for` loop is great when you know how many times to repeat. But what about "keep going until the user types quit" — where you don't know the count ahead of time? That's the `while` loop. It repeats as long as a condition stays true. Today we'll cover the basics, counting, the dreaded infinite loop, and reading input until the user stops. Let's go!

#### 1. The basic while loop

A `while` loop checks a condition, runs the body if it's true, then checks again. Three parts to remember: initialise, condition, update.

````python
count = 1                 # 1. initialise

while count <= 5:         # 2. condition
    print(count)
    count += 1            # 3. update (do NOT forget this!)
````

That `count += 1` is the most important line. Forget it and the condition never becomes false — the loop runs forever. More on that in a second.

#### 2. Counting down

Same idea, decreasing instead of increasing.

````python
n = 5
while n > 0:
    print(n, end=" ")
    n -= 1
print("-> Go!")
````

#### 3. The infinite loop trap

If the condition never turns false, the loop never ends. The usual cause is forgetting to update the variable. (We won't run an infinite loop here — just know that missing the update line is the #1 cause.)

````python
# SAFE demo of the idea: this DOES end because we update i
i = 0
while i < 3:
    print("running", i)
    i += 1     # remove this line and it never stops!
````

#### 4. while with a break — read input until 'quit'

The most common real use: keep asking until the user decides to stop. `break` exits the loop immediately. (In a video you'd type answers live; here we simulate a list of inputs.)

````python
commands = ["hello", "help", "quit", "never reached"]

for cmd in commands:                 # stand-in for real input()
    if cmd == "quit":
        print("Goodbye!")
        break
    print("You typed:", cmd)
````

In a real program that would look like the snippet below — kept here as text so the notebook doesn't wait for input:

```python
while True:
    cmd = input('Enter a command (quit to stop): ')
    if cmd == 'quit':
        break
    print('You typed:', cmd)
```

#### 5. while vs for — which to use?

- Use **for** when you know how many times / you're looping over a sequence.
- Use **while** when you repeat until some condition changes and don't know the count in advance.

#### Quick recap

- `while condition:` repeats as long as the condition is true.
- Remember the three parts: initialise, condition, update.
- Forgetting the update line causes an infinite loop.
- `while True:` + `break` is the classic "loop until told to stop" pattern.
- Use `for` for known counts, `while` for "until something happens".
