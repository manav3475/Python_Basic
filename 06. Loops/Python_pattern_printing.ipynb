### Pattern Printing Programs in Python

Pattern programs — pyramids, triangles, number grids — are a rite of passage for every Python beginner, and a favourite in interviews and college assignments. They're really just nested loops plus careful control of spaces and newlines. Today we'll build a right triangle, an inverted one, a number pyramid, and a proper star pyramid. Let's go!

#### 1. Right-angled star triangle

Row `i` prints `i` stars. Outer loop = rows, inner loop = stars in that row.

```python
rows = 5

for i in range(1, rows + 1):
    for j in range(i):
        print("*", end="")
    print()
```

#### 2. Inverted right triangle

Count the rows *down* so each line has one fewer star.

```python
rows = 5

for i in range(rows, 0, -1):
    for j in range(i):
        print("*", end="")
    print()
```

#### 3. Number triangle

Print the column number instead of a star.

```python
rows = 5

for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

#### 4. A centered star pyramid

The trick to a pyramid is the spaces. Each row prints some leading spaces, then the stars. Spaces shrink as stars grow.

```python
rows = 5

for i in range(1, rows + 1):
    spaces = rows - i
    stars = 2 * i - 1
    print(" " * spaces + "*" * stars)
```

The two formulas are the whole secret: leading spaces = `rows - i`, and stars = `2*i - 1` (which gives the odd numbers 1, 3, 5, 7... that make a symmetric pyramid).

#### 5. How to approach any pattern

1. Count the rows — that's your outer loop.
2. For each row, work out how many spaces and how many symbols.
3. Print spaces first, then symbols, then a newline.

Almost every pattern is a variation of those three steps.

#### Quick recap

- Patterns = nested loops + control of spaces and newlines.
- Outer loop = rows; inner loop(s) = what's on each row.
- `print(char, end="")` stays on the line; `print()` ends the row.
- Pyramids need leading spaces: spaces `= rows - i`, stars `= 2*i - 1`.
- Approach: count rows, then figure spaces + symbols per row.
