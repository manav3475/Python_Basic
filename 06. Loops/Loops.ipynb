### Loop control Statement

### 1. Break - The break statement kills the loop immediately. Once encountered, Python exits the loop block and continues running the rest of your program.

### 2. continue - The continue statement acts like a "skip" button. It drops whatever code is left below it inside the loop for that specific turn, then restarts the loop cycle with the next item. 

```python
i = 1
while i < 6:   
  print(i)     
  if (i == 3):   
    break
  i += 1
```

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

```python
i = 1
while i <= 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```

### FOR Loop

```python
l1 = ["Raj","Manan","Jayraj",88,77,44,566]

for i in l1:
    print(i)
```

```python
for i in l1:
    if i == 77:
        break
    print(i)
   
```

```python
for i in l1:
    if i == 77:
       continue
    print(i)
   
```

```python
for i in l1:
    if i == 77:
       continue
    print(i)
else:
    print("End of list")
```
