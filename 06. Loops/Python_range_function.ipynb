### The Python range() Function

In the last video we looped over a list. But what if you just want to repeat something 5 times, or count from 1 to 100? That's exactly what `range()` is for — it generates a sequence of numbers on demand. It's the engine behind most `for` loops in Python. Today we'll cover all three ways to call it, counting up, counting down, and a common gotcha. Let's go!

#### 1. range(stop) — the simplest form

Give it one number and it counts from 0 up to *but not including* that number.

```python
for i in range(5):
    print(i)

# prints 0 1 2 3 4  -> five numbers, starting at 0
```

Notice it stops **before** 5. `range(5)` gives you five values: 0, 1, 2, 3, 4. That "up to but not including" rule catches every beginner at least once.

#### 2. range(start, stop) — pick where to begin

Two numbers: the first is where to start, the second is where to stop (still exclusive).

```python
for i in range(1, 6):
    print(i)

# 1 2 3 4 5  -> starts at 1, stops before 6
```

#### 3. range(start, stop, step) — jump by any amount

A third number is the step — how much to jump each time. Great for even numbers, odd numbers, or any interval.

```python
# every 2nd number
for i in range(0, 11, 2):
    print(i, end=" ")
print()

# odd numbers
for i in range(1, 10, 2):
    print(i, end=" ")
```

#### 4. Counting down with a negative step

Make the step negative to count backwards — perfect for countdowns.

```python
for i in range(5, 0, -1):
    print(i, end=" ")
print("-> Lift off!")
```

#### 5. range() with len() to get positions

A common pattern: loop over the *positions* of a list using `range(len(...))` when you need the index, not just the value.

```python
names = ["Asha", "Ravi", "Meera"]

for i in range(len(names)):
    print(i, "->", names[i])
```

Tip: if you need both the index and the value, there's an even cleaner tool called `enumerate()` — we'll cover it in an upcoming video.

#### 6. range() is lazy (a quick note)

`range()` doesn't build the whole list in memory — it produces numbers one at a time. To actually see them as a list, wrap it in `list()`.

```python
print(range(5))          # range(0, 5)  -> not a list
print(list(range(5)))    # [0, 1, 2, 3, 4]
```

#### Quick recap

- `range(stop)` counts from 0 up to but NOT including stop.
- `range(start, stop)` chooses where to begin.
- `range(start, stop, step)` jumps by step; a negative step counts down.
- `range(len(seq))` gives you index positions.
- `range()` is lazy — wrap in `list()` to see the numbers.
