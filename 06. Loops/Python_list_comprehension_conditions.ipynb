### List Comprehension with Conditions

A plain comprehension transforms every item. But often you want to *filter* — keep only some items — or choose between two values. That's where conditions come in. There are two different spots a condition can go, and mixing them up is the #1 confusion. Today we'll clear that up completely. Let's go!

#### 1. Filtering with `if` at the END

Put an `if` after the loop to *keep* only items that pass the test. Items that fail are dropped.

```python
nums = range(10)

# keep only even numbers
evens = [n for n in nums if n % 2 == 0]
print(evens)     # [0, 2, 4, 6, 8]
```

Read it: *"give me `n` for each `n` in nums **if** n is even"*. The trailing `if` is a filter.

#### 2. A word filter

Same idea on strings — keep words longer than 4 letters.

```python
words = ["hi", "python", "code", "beginner", "fun"]

long_words = [w for w in words if len(w) > 4]
print(long_words)   # ['python', 'beginner']
```

#### 3. Filter AND transform together

You can transform the kept items at the same time — square only the even numbers.

```python
result = [n * n for n in range(10) if n % 2 == 0]
print(result)   # [0, 4, 16, 36, 64]
```

#### 4. if-else at the FRONT (this is different!)

Here's the part that confuses everyone. If you want to *choose between two values* (not filter), the `if-else` goes **before** the loop, as part of the expression.

```python
nums = range(6)

# label each number, keep them ALL
labels = ["even" if n % 2 == 0 else "odd" for n in nums]
print(labels)
```

#### 5. The rule to remember

- `if` only (a filter) goes at the **end**: `[x for x in items if cond]`.
- `if-else` (a choice) goes at the **front**: `[a if cond else b for x in items]`.

Filter = end. Choose a value = front. That single rule clears up all the confusion.

```python
# side by side
nums = [1, 2, 3, 4, 5, 6]

print([n for n in nums if n > 3])                 # filter  -> [4, 5, 6]
print(["big" if n > 3 else "small" for n in nums]) # choose  -> keeps all 6
```

#### Quick recap

- Filtering `if` goes at the END: `[x for x in items if condition]`.
- Choosing `if-else` goes at the FRONT: `[a if cond else b for x in items]`.
- Filter drops items; if-else keeps them all but picks a value.
- You can filter and transform in the same comprehension.
- Rule of thumb: filter = end, choose = front.
