### Infinite Loops & How to Avoid Them

An infinite loop is a loop that never stops — it just keeps running until you force-quit the program. Sometimes that's a bug that freezes your code; sometimes it's actually the design you want. Today we'll see what causes them, the intentional `while True` pattern with a clean exit, and the two classic bugs — the missing update and the off-by-one — so you can spot them fast. Let's go!

#### 1. What causes an infinite loop

A `while` loop keeps going as long as its condition is true. If the condition can *never* become false, it loops forever. The most common cause: forgetting to update the variable in the condition. (We won't actually run one — here's the buggy shape as text.)

```python
i = 0
while i < 5:
    print(i)
    # BUG: no i += 1, so i stays 0 forever
```

#### 2. The fix — always update the condition variable

Add the update line and the loop ends properly.

````python
i = 0
while i < 5:
    print(i, end=" ")
    i += 1        # the fix: i moves toward 5
print("-> done")
````

#### 3. Intentional infinite loops — `while True` + break

Sometimes you *want* to loop forever and exit only when something happens. `while True` with a `break` is the standard, readable pattern for this. (Simulated with a list so the notebook doesn't hang waiting for input.)

````python
inputs = ["3", "7", "0", "never reached"]
total = 0

for value in inputs:               # stands in for input()
    if value == "0":
        print("Zero entered - stopping.")
        break
    total += int(value)

print("Total:", total)
````

The real version would be:

```python
while True:
    value = input('Enter a number (0 to stop): ')
    if value == '0':
        break
    total += int(value)
```

#### 4. Classic bug: the missing update

We covered it above — it's worth repeating because it's the #1 cause. Any `while` loop should have something in the body that eventually makes the condition false.

#### 5. Classic bug: off-by-one

Not infinite, but the sibling bug — running one time too many or too few because of `<` vs `<=` or a wrong start. Watch the difference.

````python
# want to print 1..5

print("with < 5 :", end=" ")
for i in range(1, 5):      # stops at 4 - one too few!
    print(i, end=" ")
print()

print("with < 6 :", end=" ")
for i in range(1, 6):      # correct - 1 through 5
    print(i, end=" ")
````

#### 6. A safety tip

When writing a `while` loop, add a safety counter while developing, so a bug can't hang your machine. Remove it once you're confident.

````python
guard = 0
x = 10
while x != 3:
    x -= 1
    guard += 1
    if guard > 1000:          # safety valve during development
        print("Safety stop - possible infinite loop!")
        break
print("x reached", x, "in", guard, "steps")
````

#### Quick recap

- An infinite loop never stops because its condition never becomes false.
- #1 cause: forgetting to update the variable in the condition.
- `while True:` + `break` is the *intentional* infinite loop with a clean exit.
- Off-by-one bugs come from `<` vs `<=` or a wrong start value.
- During development, a safety counter can stop a runaway loop.
