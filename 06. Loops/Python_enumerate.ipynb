### Looping with enumerate() in Python

Very often you want both the item AND its position while looping. Beginners reach for `range(len(...))` and index in — but Python has a cleaner, more readable tool made exactly for this: `enumerate()`. It hands you the index and the value together. Today we'll see the messy way, the clean way, changing the start number, and a practical example. Let's go!

#### 1. The messy way first

Here's how people get the index without enumerate — it works, but it's clunky.

```python
names = ["Asha", "Ravi", "Meera"]

for i in range(len(names)):
    print(i, names[i])
```

#### 2. The clean way — enumerate()

`enumerate()` gives you `(index, value)` on each turn. Unpack them into two variables and you're done — no `len()`, no indexing.

```python
names = ["Asha", "Ravi", "Meera"]

for index, name in enumerate(names):
    print(index, name)
```

#### 3. Start counting from 1 (or any number)

By default enumerate starts at 0. Pass `start=` to begin somewhere else — perfect for numbered lists shown to humans.

```python
items = ["Milk", "Eggs", "Bread"]

for num, item in enumerate(items, start=1):
    print(f"{num}. {item}")
```

#### 4. A practical example — finding positions

enumerate makes it easy to record *where* something is while you scan a list.

```python
scores = [42, 91, 68, 91, 55]
top_positions = []

for i, score in enumerate(scores):
    if score == 91:
        top_positions.append(i)

print("91 found at positions:", top_positions)
```

#### 5. It works on strings and any sequence too

Anything you can loop over, you can enumerate.

```python
for i, ch in enumerate("code"):
    print(i, ch)
```

#### Quick recap

- `enumerate(seq)` gives `(index, value)` pairs while looping.
- Unpack them: `for i, item in enumerate(seq):` — no more `range(len(...))`.
- Use `start=1` (or any number) to change where the count begins.
- Great for numbered menus and recording positions.
- Works on lists, strings, tuples — any sequence.
