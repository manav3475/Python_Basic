### Looping Through Strings in Python

A string is really just a sequence of characters — which means you can loop over it, one character at a time, exactly like a list. This unlocks a ton of everyday tasks: counting letters, checking for vowels, building new text. Today we'll iterate over a string, count vowels, loop with indexes, and reverse a string. Let's go!

#### 1. A string is a sequence — loop right over it

A `for` loop hands you each character in turn.

```python
word = "python"

for ch in word:
    print(ch)
```

#### 2. Counting something — vowels in a word

A classic beginner exercise: walk through each character and count how many are vowels.

```python
word = "programming"
vowels = "aeiou"
count = 0

for ch in word:
    if ch in vowels:
        count += 1

print("Vowels:", count)
```

#### 3. Looping with the index too

If you need each character's position, loop over `range(len(...))`.

```python
word = "code"

for i in range(len(word)):
    print(i, "->", word[i])
```

#### 4. Building a new string as you go

Strings can't be changed in place, but you can build a brand-new one. Here we keep only the consonants.

```python
word = "education"
vowels = "aeiou"
result = ""

for ch in word:
    if ch not in vowels:
        result += ch

print(result)     # dctn
```

#### 5. Reversing a string

You can reverse by looping, but Python's slice trick `[::-1]` is the neat one-liner. Both are worth knowing.

```python
word = "hello"

# the loop way
rev = ""
for ch in word:
    rev = ch + rev
print(rev)

# the slice way (one line)
print(word[::-1])
```

#### 6. Counting each character with a dictionary (bonus)

Loop over the string and tally how often each character appears.

```python
word = "banana"
freq = {}

for ch in word:
    freq[ch] = freq.get(ch, 0) + 1

print(freq)     # {'b': 1, 'a': 3, 'n': 2}
```

#### Quick recap

- A string is a sequence, so `for ch in word:` gives each character.
- Count things by combining a loop with an `if` and a counter.
- Use `range(len(word))` when you need the index.
- Build a new string with `+=` (strings can't be changed in place).
- Reverse quickly with `word[::-1]`.
