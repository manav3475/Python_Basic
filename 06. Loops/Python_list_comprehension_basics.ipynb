### List Comprehension Basics in Python

You'll often build a new list by looping and appending. Python has a shorter, faster, and very Pythonic way to do the exact same thing in one line — the list comprehension. Once it clicks, you'll use it everywhere. Today we'll convert a normal loop into a comprehension, transform values, and see when NOT to use it. Let's go!

#### 1. The long way — loop and append

Here's the familiar pattern: make an empty list, loop, append.

```python
squares = []
for x in range(10):
    squares.append(x * x)

print(squares)
```

#### 2. The same thing as a list comprehension

A comprehension packs that whole pattern into one line: `[expression for item in iterable]`.

```python
squares = [x * x for x in range(10)]
print(squares)
```

Read it left to right: *"give me `x*x` for each `x` in range(10)"*. The expression on the left is what goes into the new list.

#### 3. Transforming values

The expression can be anything — a calculation, a method call, a conversion.

```python
# double each number
nums = [1, 2, 3, 4]
print([n * 2 for n in nums])

# uppercase each word
words = ["python", "is", "fun"]
print([w.upper() for w in words])

# string lengths
print([len(w) for w in words])
```

#### 4. It works on any iterable

Strings, ranges, existing lists — anything you can loop over.

```python
# character list from a string
print([ch for ch in "hello"])

# convert a list of strings to ints
price_strs = ["10", "25", "40"]
print([int(p) for p in price_strs])
```

#### 5. When NOT to use one

Comprehensions are for building a list. If you're just doing side effects (like printing), or the logic is complex with several steps, a normal loop is clearer. Readability wins.

```python
# DON'T do this just to print - a plain loop is clearer
# [print(x) for x in range(3)]   # builds a useless list of None

# DO use a comprehension when you actually want the list
result = [x * x for x in range(3)]
print(result)
```

#### Quick recap

- Syntax: `[expression for item in iterable]`.
- It replaces the empty-list + loop + append pattern in one line.
- The left expression can transform each item (`n*2`, `w.upper()`, `int(p)`).
- Works on any iterable: lists, strings, ranges.
- Use a normal loop for side effects or complex multi-step logic.
