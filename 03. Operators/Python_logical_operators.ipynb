### Python Logical Operators: and, or, not

Real decisions usually depend on more than one condition at once: the user must be logged in **and** an admin; the ticket is free **or** the user has a pass. Python's three logical operators — `and`, `or`, `not` — let you combine `True`/`False` values into a single answer. We'll cover all three, build a real eligibility check, and see the clever short-circuit behaviour that makes them do double duty. Let's go!

#### 1. The three operators

- `and` -> True only if **both** sides are True
- `or`  -> True if **at least one** side is True
- `not` -> flips True to False and back

```python
print(True and True)     # True
print(True and False)    # False
print(False or True)     # True
print(False or False)    # False
print(not True)          # False
print(not False)         # True
```

#### 2. Combining real conditions

The real power shows when the operands are comparisons. Here's a login eligibility check.

```python
age = 22
has_id = True

# must be 18+ AND have an ID
can_enter = age >= 18 and has_id
print("Can enter:", can_enter)

is_weekend = False
is_holiday = True
# day off if it is the weekend OR a holiday
print("Day off:", is_weekend or is_holiday)
```

#### 3. Short-circuit evaluation

Python is lazy in a smart way. With `and`, if the first part is already `False`, it never even checks the second — the answer can't change. With `or`, if the first part is `True`, it stops right there. This is called short-circuiting.

```python
def check():
    print("check() was called")
    return True

# first is False, so Python never calls check()
print(False and check())    # just False, no "check() was called"

# first is True, so Python never calls check()
print(True or check())      # just True
```

#### 4. and / or return a *value*, not just True/False

Here's the trick that makes logical operators so handy: they return one of the actual operands. `or` is perfect for supplying a fallback/default value.

```python
name = ""
# empty string is falsy, so or falls through to the default
display = name or "Guest"
print(display)             # Guest

name = "Manav"
print(name or "Guest")     # Manav
```

#### 5. Operator order

When you mix them, `not` runs first, then `and`, then `or`. Use parentheses when you want to be crystal clear about your intent.

```python
print(True or False and False)     # True  -> and binds tighter than or
print((True or False) and False)   # False -> parentheses change it
```

#### Quick recap

- `and` -> both must be True; `or` -> at least one; `not` -> flips it.
- Combine comparisons to check several conditions at once.
- Short-circuiting: `and` stops on the first False, `or` on the first True.
- `or` returns the first truthy value -> great for defaults (`name or "Guest"`).
- Precedence: `not` -> `and` -> `or`; parentheses make intent clear.
