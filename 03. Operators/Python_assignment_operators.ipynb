### Python Assignment Operators

You already know the basic `=` that puts a value into a variable. But Python has a whole family of *augmented* assignment operators — `+=`, `-=`, `*=`, and friends — that let you update a variable based on its own current value, with less typing and cleaner code. Today we'll cover the basic assignment, all the shortcut operators, and see how they work on strings and lists too. Let's go!

#### 1. Basic assignment `=`

The `=` sign stores the value on the right into the name on the left.

```python
x = 10
print(x)

# reassign to a new value
x = 25
print(x)
```

#### 2. Augmented assignment — update in place

Very often you want to change a variable using its current value, like adding to a running total. `total = total + 5` works, but `total += 5` says the same thing more compactly.

```python
test = 4

print(test)

test = test + 1

print(test)
```

```python
total = 100

total += 50     # same as: total = total + 50
print(total)    # 150

total -= 30     # subtract
print(total)    # 120

total *= 2      # multiply
print(total)    # 240
```

#### 3. The full set

Every arithmetic operator has a matching augmented version.

```python
n = 20

n /= 4      # true division  -> 5.0 n = n / 4
print(n)

n = 20
n //= 3     # floor division -> 6 n = n // 3
print(n)

n = 20
n %= 7      # modulo        -> 6 n = n % 7
print(n)

n = 2
n **= 5     # exponent      -> 32 n = n ** 5
print(n)
```

#### 4. They work on strings and lists too

`+=` isn't just for numbers. On a string it appends; on a list it extends. This is a handy way to build things up piece by piece.

```python
msg = "Hello"
msg += " World"     # append text
print(msg)          # Hello World

nums = [1, 2, 3]
nums += [4, 5]      # extend the list
print(nums)         # [1, 2, 3, 4, 5]
```

#### 5. Multiple assignment (bonus)

Python also lets you assign several variables in one line — the same value to many names, or unpack several values at once.

```python
a = b = c = 0        # same value to all three
print(a, b, c)

x, y, z = 1, 2, 3    # unpack three at once
print(x, y, z)

x, y = y, x          # swap without a temp variable
print(x, y)

a = 12,44,"Man"
print(a)
```

#### Quick recap

- `=` stores a value into a variable.
- Augmented operators update in place: `+=  -=  *=  /=  //=  %=  **=`.
- `x += 5` is shorthand for `x = x + 5` — less typing, clearer intent.
- `+=` also appends to strings and extends lists.
- Assign many at once (`a = b = 0`) or unpack (`x, y = 1, 2`) and swap (`x, y = y, x`).
