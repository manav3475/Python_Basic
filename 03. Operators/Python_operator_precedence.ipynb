### Python Operator Precedence & Associativity

When an expression mixes several operators — `2 + 3 * 4 ** 2` — Python needs rules to decide what runs first. Those rules are **precedence** (which operator wins) and **associativity** (left-to-right or right-to-left when operators tie). Get them wrong and your math is silently off. We'll walk the important order, see where people slip, and learn the one habit that saves you every time: parentheses. Let's go!

#### 1. Precedence — some operators run first

Just like BODMAS/PEMDAS in school: `**` runs before `*` and `/`, which run before `+` and `-`.

```python
print(2 + 3 * 4)       # 14, not 20  -> * before +
print(2 ** 3 * 2)      # 16          -> ** before *
print(10 - 2 + 3)      # 11          -> + and - left to right
print(4 - (2*2) + (2-1))  # 3          -> parentheses first
```

#### 2. A rough order, high to low

From strongest to weakest binding:

1. `()` parentheses
2. `**` exponent
3. `*  /  //  %`
4. `+  -`
5. comparisons: `==  !=  >  <  >=  <=`
6. `not`
7. `and`
8. `or`

So arithmetic happens before comparisons, which happen before the logical operators.

```python
# arithmetic first, then comparison, then logic
result = 5 + 3 > 6 and 2 * 2 == 4
# (8 > 6) and (4 == 4) -> True and True
print(result)          # True
```

#### 3. Associativity — direction when operators tie

When two operators have the same precedence, associativity decides the direction. Most are left-to-right, but `**` is special: it goes **right-to-left**.

```python
print(2 ** 3 ** 2)     # 512, not 64  -> right to left: 2 ** (3 ** 2) = 2 ** 9
print((2 ** 3) ** 2)   # 64           -> forced left with parentheses

print(100 // 5 // 2)   # 10           -> left to right: (100 // 5) // 2
```

#### 4. The gotcha this prevents

Precedence bugs are silent — the code runs, it's just wrong. Here the intent was to average two numbers, but the missing parentheses divide only the second one.

```python
a, b = 10, 20

wrong = a + b / 2      # b/2 first -> 10 + 10 = 20
right = (a + b) / 2    # 30 / 2    -> 15.0

print("wrong:", wrong)
print("right:", right)
```

#### 5. The habit that always wins

You don't have to memorize the whole table. When an expression mixes operators, wrap the part you mean in parentheses. It removes ambiguity and makes your intent obvious to the next reader.

```python
# hard to read at a glance
x = 2 + 3 * 4 - 1

# intent is crystal clear
x = 2 + (3 * 4) - 1
print(x)   # 13
```

#### Quick recap

- **Precedence** decides which operator runs first: `()` -> `**` -> `* / // %` -> `+ -` -> comparisons -> `not` -> `and` -> `or`.
- Arithmetic runs before comparisons, which run before logical operators.
- **Associativity** breaks ties: most operators go left-to-right, but `**` goes right-to-left.
- Precedence bugs are silent and wrong — the classic is `(a + b) / 2` vs `a + b / 2`.
- When in doubt, add parentheses for clarity.
