### Python Membership Operators: in / not in

How do you ask *"is this item inside that collection?"* Python makes it read like plain English with two membership operators: `in` and `not in`. They work on lists, strings, tuples, sets, and dictionaries — and they return a simple `True` or `False`. We'll try each one and point out the one surprise with dictionaries. Let's go!

#### 1. `in` with a list

Ask whether a value appears in a list.

```python
fruits = ["apple", "banana", "mango"]

print("apple" in fruits)      # True
print("grape" in fruits)      # False
print("grape" not in fruits)  # True
```

#### 2. `in` with a string

On a string, `in` checks for a **substring** — a very common way to test if some text contains a piece.

```python
sentence = "learning python is fun"

print("python" in sentence)   # True
print("java" in sentence)     # False
print("Python" in sentence)   # False -> case sensitive
```

#### 3. `in` with a dictionary checks KEYS, not values

This is the one to remember: on a dictionary, `in` looks at the **keys** by default, not the values.

```python
student = {"name": "Manav", "age": 25}

print("name" in student)      # True  -> "name" is a key
print("Manav" in student)     # False -> "Manav" is a value, not a key

# to search values explicitly:
print("Manav" in student.values())   # True
```

#### 4. Works on tuples and sets too

Membership works across collections. On sets it's especially fast, which is why sets are great for repeated "have I seen this?" checks.

```python
coords = (10, 20, 30)
print(20 in coords)          # True
print(30 not in coords)       # True

seen = {"a", "b", "c"}
print("b" in seen)           # True
print("z" not in seen)       # True
```

#### 5. A practical use — validating input

Membership shines when checking a value against a set of allowed options.

```python
valid_commands = ["start", "stop", "pause"]
command = "checking"

if command in valid_commands:
    print("Running:", command)
else:
    print("Unknown command")
```

#### Quick recap

- `in` -> True if the item is found; `not in` -> the opposite.
- On a string, `in` checks for a **substring** (case sensitive).
- On a dictionary, `in` checks the **keys**; use `.values()` to search values.
- Works on lists, tuples, and sets — sets are the fastest for membership tests.
- Great for validating input against a list of allowed options.
