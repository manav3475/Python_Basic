### Python Comparison Operators

Programs constantly ask yes/no questions: is this score high enough? are these two names the same? is the price within budget? Comparison operators are how you ask. Python has six — `==`, `!=`, `>`, `<`, `>=`, `<=` — and they always hand back a simple `True` or `False`. Today we'll run through all six, compare numbers *and* text, chain them the Pythonic way, and look at two gotchas that catch every beginner. Let's go!

#### 1. The six operators

Each comparison evaluates to a Boolean — `True` or `False`.

```python
a = 10
b = 3

print("a == b :", a == b)   # equal?
print("a != b :", a != b)   # not equal?
print("a >  b :", a > b)
print("a <  b :", a < b)
print("a >= b :", a >= b)
print("a <= b :", a <= b)
```

Notice `==` (two equals signs) asks *"are these equal?"*, while a single `=` **assigns** a value. Mixing them up is one of the most common beginner bugs.

#### 2. Comparing strings

Strings compare alphabetically (technically by Unicode code point). Uppercase letters come before lowercase, which surprises people.

```python
print("apple" == "apple")    # True
print("apple" == "Apple")    # False -> case matters
print("apple" < "banana")    # True  -> a comes before b
print("Zebra" < "apple")     # True  -> uppercase Z (90) < lowercase a (97)
```

#### 3. Chaining comparisons

Python lets you chain comparisons in one clean expression, just like in math. `18 <= age <= 60` reads exactly how you'd say it out loud.

```python
age = 25
print(18 <= age <= 60)     # True -> is age between 18 and 60?

score = 95
print(0 <= score <= 100)   # True -> valid percentage?
```

#### 4. Gotcha: floats don't always compare how you'd expect

Because computers store decimals in binary, some values are tiny bit off. The classic example: `0.1 + 0.2` is not exactly `0.3`. Run this and watch.

```python
print(0.1 + 0.2)                 # 0.30000000000000004
print(0.1 + 0.2 == 0.3)          # False (!)

# the safe way to compare floats:
import math
print(math.isclose(0.1 + 0.2, 0.3))   # True
```

#### 5. Gotcha: you can't order different types

Checking equality across types is fine (it's just `False`), but asking whether a string is *greater than* a number makes no sense — Python raises a `TypeError`. Let's trigger it on purpose.

```python
print(5 == "5")     # False -> different types, safe
print(5 < "5")      # TypeError: cannot compare int and str
```

#### Quick recap

- Six operators: `==  !=  >  <  >=  <=`, each returning `True` or `False`.
- `==` compares values; a single `=` assigns — don't mix them up.
- Strings compare alphabetically by Unicode; case matters.
- Chain naturally: `0 <= score <= 100`.
- Compare floats with `math.isclose()`, not `==`.
- Ordering different types (`int < str`) raises `TypeError`.
