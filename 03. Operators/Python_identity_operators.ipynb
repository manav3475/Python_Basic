### Python Identity Operators: is / is not

`is` and `==` look similar and beginners mix them up constantly — but they ask two different questions. `==` asks *"do these have the same value?"* while `is` asks *"are these the exact same object in memory?"* Getting this right matters, especially when checking for `None`. We'll cover the difference, the correct way to test for `None`, and a famous small-integer gotcha. Let's go!

#### 1. `is` vs `==`

`==` compares values. `is` compares identity — whether two names point to the very same object. The built-in `id()` shows an object's identity.

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a
d = a.copy()

print(a == b)    # True  -> same contents
print(a is b)    # False -> two separate list objects
print(a is c)    # True  -> c points to the same object as a
print(d is a)

print(id(a), id(b), id(c), id(d))  # a and c have the same id, b and d have different ids
```

So `a` and `b` hold equal values but are different objects, while `c` is literally the same object as `a`. Change one through `c` and `a` sees it too.

```python
c.append(4)
print(a)     # [1, 2, 3, 4] -> a changed because c IS a

d.append(7)
print(d)     # [1, 2, 3, 7] -> d changed but a remains the same
print(a)     # [1, 2, 3, 4] -> a remains the same because d is a copy of a
```

#### 2. The right way to check for None

`None` is a single, unique object in Python. The convention — and the recommended style — is to test it with `is`, never `==`.

```python
value = None

print(value is None)       # True  -> correct, Pythonic
print(value is not None)   # False

result = 42
print(result is not None)  # True
```

#### 3. The small-integer caching gotcha

This one surprises everyone. To save memory, Python pre-creates and reuses small integers (roughly -5 to 256). So `is` can accidentally look `True` for small numbers and `False` for bigger ones — even though `==` is correct in both cases. **Never use `is` to compare values.**

```python
x = 256
y = 256
print(x is y)     # True  -> 256 is cached and reused

x = 257
y = 257
print(x is y)     # False -> outside the cache, two objects

print(257 == 257) # True  -> value comparison is always correct
```

#### Quick recap

- `==` compares **values**; `is` compares **identity** (same object in memory).
- `id()` reveals an object's identity.
- Two equal lists can be `==` True but `is` False.
- Always test `None` with `is None` / `is not None`.
- Small ints (-5..256) are cached, so `is` on numbers is unreliable — use `==` for values.
