### Python Arithmetic Operators

Every program does math at some point — adding up a bill, splitting people into groups, checking if a number is even. Python gives you seven arithmetic operators to do it: `+`, `-`, `*`, `/`, `//`, `%`, and `**`. Most feel obvious, but three of them — true division, floor division, and modulo — trip up almost every beginner. Today I'll walk through all seven with clear examples, and really slow down on floor division and modulo so they finally click. Let's go!

#### 1. The four you already know: `+`  `-`  `*`  `/`

Addition, subtraction, and multiplication behave exactly like you'd expect. Division has one surprise we'll get to in a second.

```python
a = 10
b = 3

print("add:", a + b)
print("sub:", a - b)
print("mul:", a * b)
print("div:", a / b)
```

#### 2. True division `/` always gives a float

Here's the surprise: in Python 3, the `/` operator **always** returns a float — even when the numbers divide evenly. `10 / 2` is `5.0`, not `5`.

```python
print(10 / 2)      # 5.0, not 5
print(type(10 / 2))
```

#### 3. Floor division `//` — divide and round DOWN

Floor division divides, then throws away the decimal part, giving you the whole number underneath. Think of it as *"how many whole times does b fit into a?"* — perfect for splitting things into equal groups.

```python
print(7 / 2)       # 3.5  -> true division
print(7 // 2)      # 3    -> floor division, rounds down

# 17 students, 5 per team -> how many full teams?
print(17 // 5)     # 3 full teams
```

#### 4. Modulo `%` — the REMAINDER left over

Modulo is the partner of floor division: it gives you what's *left over* after the division. `17 % 5` is `2`, because 5 fits into 17 three times (15) with 2 left over.

```python
print(17 % 5)      # 2  -> remainder

# floor division + modulo together tell the whole story:
print(17 // 5, "teams with", 17 % 5, "left over")
```

The single most common use of modulo is the **even / odd check**. A number is even when dividing by 2 leaves no remainder.

```python
num = 8
print(num % 2)            # 0 -> even
print(num % 2 == 0)       # True

num = 7
print(num % 2)            # 1 -> odd
```

#### 5. The negative-number gotcha

This one surprises people: with negative numbers, `//` rounds **down toward negative infinity**, not toward zero. So `-7 // 2` is `-4`, not `-3`. Modulo follows along to stay consistent, and in Python the result of `%` always takes the sign of the divisor.

```python
print(-7 // 2)     # -4  (rounds DOWN, not toward zero)
print(-7 % 2)      #  1  (sign follows the divisor)

# the identity that always holds:  (a // b) * b + (a % b) == a
print((-7 // 2) * 2 + (-7 % 2))   # -7
```

#### 6. Exponentiation `**` — powers

Two stars means "to the power of". `2 ** 3` is 2 × 2 × 2 = 8. It also handles roots via fractional powers, and Python's integers can grow as large as you want without overflow.

```python
print(2 ** 3)          # 8
print(5 ** 2)          # 25
print(9 ** 0.5)        # 3.0  -> square root via power of 1/2
print(2 ** 100)        # huge ints are no problem in Python
```

#### 7. Watch out: dividing by zero

Math says you can't divide by zero, and Python agrees — it stops the program with a `ZeroDivisionError`. Let's trigger it on purpose so you recognize it when it shows up.

```python
print(10 / 0)     # ZeroDivisionError: division by zero
```

The same happens with `//` and `%` against zero. Later, in the exceptions video, we'll learn how to catch this gracefully with `try` / `except` instead of crashing.

#### 8. Operator precedence — what runs first

When you mix operators, Python follows an order (like BODMAS/PEMDAS from school): `**` first, then `*  /  //  %`, then `+  -` last. When in doubt, use parentheses to make your intent obvious and your code readable.

```python
print(2 + 3 * 4)       # 14, not 20  -> * before +
print((2 + 3) * 4)     # 20          -> parentheses win
print(2 ** 3 * 2)      # 16          -> ** before *
```

#### Quick recap

- `+  -  *` behave normally; `/` (true division) **always returns a float**.
- `//` (floor division) divides and rounds **down** to a whole number.
- `%` (modulo) gives the **remainder** — the go-to tool for even/odd and cycling.
- `//` and `%` round toward negative infinity, so watch the sign with negatives.
- `**` raises to a power; a fractional power like `** 0.5` gives a root.
- Dividing by zero raises `ZeroDivisionError`.
- Precedence: `**` -> `* / // %` -> `+ -`; parentheses override everything.
