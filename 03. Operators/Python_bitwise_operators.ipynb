### Python Bitwise Operators

Every number lives in the computer as bits — 1s and 0s. Bitwise operators let you work on those individual bits directly. You won't use them every day, but they're essential for flags, permissions, low-level data, and plenty of interview questions. We'll cover all six — `&`, `|`, `^`, `~`, `<<`, `>>` — and use `bin()` to actually see what's happening. This one's a step up, so let's take it slow. Let's go!

#### 1. Seeing numbers in binary

The `bin()` function shows a number's bit pattern. That makes the bitwise operations much easier to follow.

```python
print(bin(5))     # 0b101
print(bin(3))     # 0b011

# 5 is 101, 3 is 011
```

#### 2. AND `&`, OR `|`, XOR `^`

These compare two numbers bit by bit:
- `&` -> 1 only where **both** bits are 1
- `|` -> 1 where **either** bit is 1
- `^` -> 1 where the bits **differ** (exclusive or)

```python
a = 5   # 101
b = 3   # 011

print(a & b, "->", bin(a & b))   # 1  -> 001
print(a | b, "->", bin(a | b))   # 7  -> 111
print(a ^ b, "->", bin(a ^ b))   # 6  -> 110
```

#### 3. NOT `~` — flip every bit

`~` inverts all the bits. Because of how Python stores signed numbers (two's complement), the handy shortcut to remember is `~x == -(x + 1)`.

```python
print(~5)     # -6   (same as -(5 + 1))
print(~0)     # -1
```

#### 4. Shifts `<<` and `>>`

Shifting moves bits left or right. A left shift by 1 doubles the number; a right shift by 1 halves it (dropping any remainder). Fast multiply/divide by powers of two.

```python
print(3 << 1, "->", bin(3 << 1))   # 6  -> 110  (doubled)
print(3 << 2, "->", bin(3 << 2))   # 12 -> 1100 (x4)
print(12 >> 1, "->", bin(12 >> 1)) # 6  -> 110  (halved)
```

#### 5. Real use case: flags

Bitwise ops let you pack several on/off settings into one number. Each permission is a bit; you combine them with `|` and check them with `&`.

```python
READ  = 0b001
WRITE = 0b010
EXEC  = 0b100

# grant read + write
perms = READ | WRITE
print(bin(perms))            # 0b011

# does the user have write access?
print(bool(perms & WRITE))   # True
print(bool(perms & EXEC))    # False
```

#### Quick recap

- Bitwise ops work on the individual bits of integers; `bin()` reveals them.
- `&` both-1, `|` either-1, `^` differ, `~` flip all bits.
- `~x` equals `-(x + 1)` thanks to two's complement.
- `<<` shifts left (multiply by 2 each step); `>>` shifts right (divide by 2).
- Classic use: pack on/off flags/permissions into one number.
