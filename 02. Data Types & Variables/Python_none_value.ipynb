### Python `None` — The 'No Value' Value

Sometimes you need a variable that exists but doesn't hold anything yet — no number, no text, just *nothing*. Python has a special value for exactly that: `None`. It's not zero, it's not an empty string, it's not `False` — it's Python's way of saying 'no value here.' Today I'll show what `None` really is, how to use it as a placeholder, and why it's the perfect default for a function argument. Let's go!

#### 1. What None means

`None` is a special built-in value that represents the absence of a value. There is exactly one `None` in all of Python. Its type is `NoneType`.

```python
x = None

print("Value:", x)
print("Type: ", type(x))
```

Notice printing it shows `None`, and its type is `NoneType` — a type that has only this one value.

#### None is not zero, empty, or False

This is the key thing to understand. `None` is its own distinct value. It is *not* the same as `0`, an empty string, or `False` — even though all of those feel 'empty' too.

```python
print(None == 0)        # False - None is not zero
print(None == "")       # False - None is not an empty string
print(None == False)    # False - None is not False
print(None == None)     # True  - None only equals itself
```

#### Checking for None the right way

The Pythonic way to test for `None` is with `is`, not `==`. Since there's only ever one `None`, `is` checks identity directly and is the recommended style.

```python
x = None

if x is None:
    print("x has no value yet")

y = 10
if y is not None:
    print("y has a value:", y)
```

#### 2. None as a placeholder / 'no value' marker

A common use is initializing a variable before you know its real value. You set it to `None`, then fill it in later — and you can check whether it's been set yet.

```python
result = None      # we don't have a result yet

# ... some logic runs ...
numbers = []
if numbers:
    result = max(numbers)

if result is None:
    print("No result was computed")
else:
    print("Result is:", result)
```

#### 3. None as a default function argument

This is where `None` really shines. When a function has an optional argument, `None` is the standard default — it clearly signals 'the caller didn't provide this.'

```python
def greet(name, greeting=None):
    if greeting is None:
        greeting = "Hello"          # fall back to a default
    print(greeting + ",", name)

greet("Manav")                       # uses the default greeting
greet("Manav", "Good morning")       # caller provided one
```

#### Why None and not a real default?

You might ask — why not just write `greeting="Hello"` directly? Often you can. But `None` is essential when the default needs to be a fresh mutable object like a list. Using `[]` as a default is a classic Python bug because the same list gets shared across all calls. `None` avoids it.

```python
# The safe pattern for a list default
def add_item(item, basket=[]):
    # if basket is None:
    #     basket = []          # a brand-new list every call
    basket.append(item)
    return basket

print(add_item("apple"))     # ['apple']
print(add_item("banana"))    # ['banana'] - NOT ['apple', 'banana']
print(add_item("XYZ"))
```

```python
# The safe pattern for a list default
def add_item(item, basket=None):
    if basket is None:
        basket = []          # a brand-new list every call
    basket.append(item)
    return basket

print(add_item("apple"))     # ['apple']
print(add_item("banana"))    # ['banana'] - NOT ['apple', 'banana']
```

Each call gets its own fresh list because we start from `None` and build a new `[]` inside. If we had written `basket=[]` in the signature, both calls would share one list and the second would wrongly show both items.

#### 4. Functions return None by default

One more place `None` shows up on its own — a function that doesn't return anything actually returns `None`. Even `print()` does this.

```python
def say_hi():
    print("Hi!")             # prints, but has no return statement

returned = say_hi()
print("say_hi() returned:", returned)   # None
```

That's why `x = print("hello")` gives `x` the value `None` — `print` shows text but returns nothing. A subtle point, but it explains a lot of beginner confusion.

#### Quick recap

- `None` is Python's special 'no value' value; its type is `NoneType` and there's only one of it.
- It is *not* `0`, `""`, or `False` — it only equals itself.
- Check for it with `is None` / `is not None`, not `==`.
- Use `None` as a placeholder for a value you'll fill in later.
- Use `None` as a function's default argument — especially for mutable defaults like lists.
- Functions with no `return` give back `None` automatically.

`None` is small but it's everywhere in Python. Once it clicks, a lot of the language makes more sense. That's a wrap for today!
